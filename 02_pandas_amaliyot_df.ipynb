### **19-mashq.**  Yuqoridagi df dan Toshkent viloyati maydoni O'zbekiston Respublikasining necha fozini tashkil qilishini topish

```python
# Javobni shuyerga yozing
tosh_area = data.at['Toshkent viloyati','Maydoni']
uzb_area = sum(data.Maydoni)
print(f"Toshkent viloyati maydoni: {tosh_area} kv.km")
print(f"O'zbekiston Respublikasi maydoni: {uzb_area} kv.km")
print(f"Toshkent viloyati maydoni O'zbekiston respublikasining {round(tosh_area/uzb_area*100)}% ni tashkil qiladi")
```

### **20-mashq.**  Yuqoridagi df dan Toshkent viloyati aholisi O'zbekiston Respublikasining necha fozini tashkil qilishini topish
