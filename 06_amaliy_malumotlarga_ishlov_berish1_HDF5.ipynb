#5-topshiriq
1. `aqi.h5` faylidan maʼlumotlarni oʻqish va 2025-yilga tegishli maʼlumotlarni yangi dataframeʼga saqlash.
2. `aqi.h5` faylida 2025-yilning qaysi oylari uchun maʼlumotlar bor? Shu oylar roʻyxatini ekranga chiqarish!

```python
#javobni shu yerda yozing.
with pd.HDFStore('aqi.h5', mode='w') as store:
    df_2022 = pd.read_csv('aqi_2022.csv')
    df_2023 = pd.read_csv('aqi_2023.csv')
    df_2024 = pd.read_csv('aqi_2024.csv')
    df_2025 = pd.read_csv('aqi_2025.csv')

    store.put('2022', df_2022)
    store.put('2023', df_2023)
    store.put('2024', df_2024)
    store.put('2025', df_2025)

    print(store.keys())
```

```python
with pd.HDFStore('aqi.h5', mode='r') as store:
    df_2025 = store['2025']   # 2025 yil ma'lumotlarini o'qib olish

# 'Date' ustunidan oylarni ajratib olish
df_2025['Date'] = pd.to_datetime(df_2025['Date'])   # Sana ustunini datetime ga o'tkazish
oylar = df_2025['Date'].dt.month.unique()          # Faqat noyob oylarni olish

# Natijani chiqarish
print("2025-yil uchun mavjud oylar:")
print(sorted(oylar))
```
