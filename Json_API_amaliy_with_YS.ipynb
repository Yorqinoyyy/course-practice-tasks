#4)
1. Quyida berilgan linkga `get` soʻrovini yuborish orqali `json` maʼlumotlarini `posts.json` fayliga saqlash.
  * https://jsonplaceholder.typicode.com/posts/
2. Ushbu maʼlumotlarni dataframe koʻrinishida saqlash va tasodifiy 5 ta qatorni ekranga chiqarish.

```python
url = "https://jsonplaceholder.typicode.com/posts/"
#javobni shu yerga yozing.
import requests
import pandas as pd
response = requests.get(url)
data = response.json()

import json
with open("posts.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)

#  DataFrame ko‘rinishida saqlash
df = pd.DataFrame(data)

# Tasodifiy 5 ta qatorni chiqarish
print(df.sample(5))
```
