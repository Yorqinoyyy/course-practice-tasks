#**6) APIʼlar bilan ishlash**
1. Yuqorida berilgan kriptovalyuta maʼlumotlaridan `id` ustunini roʻyxat koʻrinishida saqlash va bu roʻyxatga `coin_ids` deb nom berish.
2. https://api.coingecko.com/api/v3/simple/price?ids={coin_ids}&vs_currencies=usd havolasi orqali kriptovalyutalarning `usd` ga nisbatini oling (birinchi 100 ta maʼlumotni olish yetarli).
3. 2-shartda berilgan link orqali olingan maʼlumotlarni dataframe koʻrinishida saqlash. Bunda uning indeksi va `usd` ustuni boʻlsin. `index` -> kriptovalyutalar nomi, `usd` -> narxi

#**💡Ishora:** linkdagi `{coin_ids}` oʻzgaruvchi sifatida yozilgan. Biz ham 1-shartda saqlagan `coin_ids` oʻzgaruvchimizni shu qismda yozib ketish!


````python
#javobni shu yerda yozing.
coins_url = "https://api.coingecko.com/api/v3/coins/list"
response = requests.get(coins_url)
data = response.json()
df = pd.DataFrame(data)
df
````

````python
coin_ids = df['id'].head(100).tolist()
````

````python
price_url = f"https://api.coingecko.com/api/v3/simple/price?ids={','.join(coin_ids)}&vs_currencies=usd"
price_response = requests.get(price_url).json()

price_df = pd.DataFrame(price_response).T
price_df.columns = ['usd']

price_df.head()
````

#**15) CREATE and INSERT**
* `chinook.db` maʼlumotlar omboriga `users` jadvali yaratish. Jadval maydonlari (ustunlar):
  * `id`
  * `name`
  * `birth_date`
* Yaratilgan `users` jadvaliga quyidagi maʼlumotlarni qoʻshish:

    * ```
    [
    ('Alice', 25), ('Bob', 30), ('Charlie', 35), ('David', 28), ('Eve', 22),
    ('Frank', 40), ('Grace', 27), ('Hannah', 32), ('Ivy', 29), ('Jack', 31),
    ('Kevin', 33), ('Liam', 26), ('Mia', 24), ('Noah', 36), ('Olivia', 23),
    ('Peter', 39), ('Quinn', 37), ('Rachel', 34), ('Samuel', 41), ('Tina', 21)
]
 ```

#💡Ishora: CREATE TABLE, INSERT


````python
#javobni shu yerda yozing.
cursor.execute("""
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY,
        name TEXT,
        birth_date INTEGER
    )
""")
users = [
('Alice', 25), ('Bob', 30), ('Charlie', 35), ('David', 28), ('Eve', 22),
('Frank', 40), ('Grace', 27), ('Hannah', 32), ('Ivy', 29), ('Jack', 31),
('Kevin', 33), ('Liam', 26), ('Mia', 24), ('Noah', 36), ('Olivia', 23),
('Peter', 39), ('Quinn', 37), ('Rachel', 34), ('Samuel', 41), ('Tina', 21)
]
for user in users:
    cursor.execute("INSERT INTO users (name, birth_date) VALUES (?, ?)", user)
conn.commit()
cursor.execute("SELECT * FROM users;")
cursor.fetchall()

````

#**16) UPDATE and DELETE**
* `users` jadvalidagi Alice ismli foydalanuvchining yoshini yangilash.
* `users` jadvalidan Alice ismli foydalanuvchini o‘chirish.

````python
#javobni shu yerda yozing.
cursor.execute("UPDATE users SET birth_date = 26 WHERE name = 'Alice';")
conn.commit()
cursor.execute("SELECT * FROM users;")
cursor.fetchone()
````

````python
cursor.execute("DELETE FROM users WHERE name = 'Alice';")
conn.commit()
cursor.execute("SELECT * FROM users;")
cursor.fetchall()
````

#**17) f-string yordamida SQL komandalarini yozish**
* `customers` jadvalidan `FirstName` va `LastName` bo‘yicha qidiruv qiladigan f-string yozish.

````python
#javobni shu yerda yozing.
first_name = "Tim"
last_name = "Gauthier"
query = f"SELECT * FROM customers WHERE LOWER(FirstName) = LOWER('{first_name}') AND LOWER(LastName) = LOWER('{last_name}');"
cursor.execute(query)
result = cursor.fetchall()
print(result)
````

#**18)  AND and OR operators**

* `invoices` jadvalidan `BillingCountry` Germaniya yoki Fransiya boʻlgan va `Total` summasi 5 dan katta bo‘lgan kerakli qatorlarni tanlash.

#💡Ishora: AND, OR

````python
#javobni shu yerda yozing.
import pandas as pd
cursor.execute("SELECT * FROM invoices WHERE (BillingCountry = 'Germany' OR BillingCountry = 'France') AND Total > 5;")
result = cursor.fetchall()
print(result)
````
